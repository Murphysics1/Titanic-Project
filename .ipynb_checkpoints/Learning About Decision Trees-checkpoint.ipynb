```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Clean the Data

```python
raw_data_import= pd.read_csv('titanic.csv')
raw_data = raw_data_import.copy()
raw_data.head()
```

```python
raw_data['Name'] = raw_data['Name'].str.replace(" ","")
raw_data.head()
```

```python
raw_data['Name'] = raw_data['Name'].str.replace(".",",")
raw_data[['Last Name','Title',"First Name","Delete"]] = raw_data['Name'].str.split(',',expand=True)
raw_data.head()
```

```python
raw_data.columns
```

```python
100*raw_data.isnull().sum()/raw_data.shape[0]
```

```python
df = raw_data.drop(columns = ['Cabin','Ticket','Delete','Name','Last Name','First Name','Embarked'])
```

```python
df.columns
```

```python
cols = ['PassengerId', 'Title','Pclass', 'Sex', 'Age', 'SibSp', 'Parch',
       'Fare','Survived']
```

```python
data = df.copy()
```

```python
data = data[cols]
```

```python
#data['Sex']=data['Sex'].map({'male':1,'female':0})
```

```python
data.head()
```

```python
data["Title"].unique()
```

```python
cat = ['Title','Entries','Max','Min','Avg','Sex']
title_df = pd.DataFrame(columns = cat)

for title in data['Title'].unique():
    ent = len(data.loc[(data['Title'] == title)]['Age'])
    maxim = data.loc[(data['Title'] == title)]['Age'].max()
    minim = data.loc[(data['Title'] == title)]['Age'].min()
    avg = data.loc[(data['Title'] == title)]['Age'].mean()
    sex = data.loc[(data['Title'] == title)]['Sex'].unique()
    df_new = {'Title':title,'Entries':ent,'Max':maxim,'Min':minim,'Avg':avg,'Sex':sex}
    title_df = title_df.append(df_new, ignore_index = True)
title_df
```

```python
honorifics = {}

mr = ['Mr','Don', 'Rev','Major','Sir','Col','Capt','Jonkheer']
mrs = ['Mrs','Mme','theCountess','Dona','Lady']
miss = ['Miss','Mlle','Ms']
master = ['Master']
for key in mr:
    honorifics[key] = "Mr"
for key in miss:
    honorifics[key] = "Miss"
for key in mrs:
    honorifics[key] = "Mrs"
for key in master:
    honorifics[key] = "Master"  
#honorifics
```

```python
data["Title"] = data["Title"].map(honorifics)
```

Giving labels to all titles with both male and female representation

```python
for i,j in enumerate(data['Title']):
    if data.loc[i,'Title'] not in ['Mr','Mrs','Miss','Master']:
        if data.loc[i,'Sex'] == 'male':
            data.loc[i,'Title'] = 'Mr'
        else:
            data.loc[i,'Title'] = 'Mrs'
```

```python
data["Title"].unique()
```

```python
title_df2 = pd.DataFrame(columns = cat)

for title in data['Title'].unique():
    ent = len(data.loc[(data['Title'] == title)]['Age'])
    maxim = data.loc[(data['Title'] == title)]['Age'].max()
    minim = data.loc[(data['Title'] == title)]['Age'].min()
    avg = data.loc[(data['Title'] == title)]['Age'].mean()
    sex = data.loc[(data['Title'] == title)]['Sex'].unique()
    df_new = {'Title':title,'Entries':ent,'Max':maxim,'Min':minim,'Avg':avg,'Sex':sex}
    title_df2 = title_df2.append(df_new, ignore_index = True)
title_df2
```

### Fill In Missing Data using KNN

```python
data_nan = data_w_cat.drop(columns = ['PassengerId','Sex'])
```

```python
data_nan['Title'] = data_nan['Title'].map({'Mr':1,'Mrs':2,'Miss':3,'Master':4})
```

```python
data_nan.info()
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=5)
filled_data = imputer.fit_transform(data_nan)
```

```python
cols2 =['Title','Pclass','Age','SibSp','Parch','Fare','Survived']
```

```python
data_filled = pd.DataFrame(filled_data, columns=cols2)
```

```python
data_filled
```

```python
data_filled.info()
```

```python
data_w_cat[['Age','Fare']] = data_filled[['Age','Fare']]
```

```python
data_w_cat.isnull().sum()
```

```python
data_w_cat.info()
```

```python
data_w_cat.head()
```

```python
data_w_cat = data_w_cat.drop(columns = ['Title','PassengerId'])
data_w_cat['Sex']=data_w_cat['Sex'].map({'male':1,'female':0})
data_w_cat['Age'] = data_w_cat['Age'].round(0)
data_w_cat['Age'] = data_w_cat['Age'].astype(int)
```

```python
data_processed = data_w_cat.copy()
```

```python
data_processed
```

## Make Train and Test Dataset

## Decision Tree Model

```python
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.image as pltimg
```

```python
features = ['Pclass','Sex']
target = data_processed['Survived']
inputs = data_processed.drop(['Fare','Survived','Age','SibSp','Parch'],axis=1)
```

```python
target
```

```python
inputs
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(inputs, target)
tree_export = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(tree_export)
graph.write_png('mydecisiontree.png')

img=pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show()
```

## Explore the Data

```python
grouped_data_SC = data_processed.groupby(["Sex","Pclass"]).mean()
grouped_data_SC = grouped_data_SC.rename(columns={"Survived":"% Survived"})
grouped_data_SC["% Survived"] = grouped_data_SC["% Survived"].apply(lambda x: x*100).round(2)
grouped_data_SC = grouped_data_SC.drop(columns = ['Age','SibSp','Parch','Fare'])

data_age = data_processed.copy()
data_age['Age']=data_age['Age'].round(-1)
grouped_data_age = data_age.groupby(["Age","Pclass"]).mean()
grouped_data_age = grouped_data_age.rename(columns={"Survived":"% Survived"})
grouped_data_age["% Survived"] = grouped_data_age["% Survived"].apply(lambda x: x*100).round(2)
grouped_data_age = grouped_data_age.drop(columns = ['SibSp','Parch','Fare'])
```
